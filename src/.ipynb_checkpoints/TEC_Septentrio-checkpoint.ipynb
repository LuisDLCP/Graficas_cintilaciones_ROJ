```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
direct = "/home/luis/Desktop/Proyects_Files/LISN/GPSs/Tareas/Comparacion_TEC/data/"
df1 = pd.read_csv(direct + "ljic_200927.dat", header=None, sep="\t")
df1.head(3)
```

```python
# Extraccion de ciertas columnas
for i in df1.columns:
    if i>3:
        m = (i-4)%11
        if m == 1: del df1[i]
        elif m == 2: del df1[i]
        elif m == 7: del df1[i]
        elif m == 8: del df1[i]
        elif m == 9: del df1[i]
        elif m == 10: del df1[i]

#df1.head()
df1
```

```python
df1.columns = np.arange(len(df1.columns)) # Rename the columns 
df1.head(2)
```

```python
# Creacion de un nuevo dataframe
df2 = pd.DataFrame(columns=range(8))

# Se llenan el resto de valores
for i in range(len(df1)):
    for h in range(df1.iloc[i,3]):
        m = df1.iloc[i,[0,1,2,4+5*h,5+5*h,6+5*h,7+5*h,8+5*h]]
        m.index = range(len(m))
        df2 = df2.append(m)

df2.index = range(len(df2))

# Drop na values 
for i in range(4):
    df2[4+i] = df2[4+i].astype("str").str.strip().astype("float")
df2.dropna(inplace = True)

# Change type object to double 
for i in range(3):
    df2[i] = df2[i].astype("int")

# Reindex 
df2.index = range(len(df2)) 

# Rename columns 
df2.columns = ["YY", "DOY", "SOD", "PRN", "TEC-45", "TEC-30", "TEC-15", "TEC"]

df2
```

```python
# Split dataframe by TEC types
df2 = pd.melt(df2,id_vars=["YY","DOY","SOD","PRN"],var_name="sTEC_type",value_name="sTEC_value")

# Restar un valor de acuerdo al grupo 
df3 = df2.groupby("sTEC_type")
df2["SOD"] = pd.concat([df3.get_group("TEC-45")["SOD"]-45,df3.get_group("TEC-30")["SOD"]-30,df3.get_group("TEC-15")["SOD"]-15,df3.get_group("TEC")["SOD"]])

# Delete "sTEC_type" column 
del df2["sTEC_type"]

# Rename a column 
df2.rename(columns={"sTEC_value":"sTEC"},inplace=True)

df2
```

```python
# Extract only 'GPS' PRNs

# Filter PRN values which begins with "G" letter
mask = df2["PRN"].str.contains("G")
df3 = df2[mask]

# Eliminate "G" letter and converts to int
df3["PRN"] = df3["PRN"].str.replace("G","").astype("int")

# Sort values by "PRN" and "SOD" 
df3 = df3.sort_values(["PRN","SOD"])

# Reindex in ascendent order 
df3.index = range(len(df3))

df3
```

```python
# Convert to datetime data type 
def change2datetime(row):
    yy = int(row[0])
    doy = int(row[1])
    sod = int(row[2])
    
    if sod < 0: 
        doy = doy -1
        sod = 60*60*24+sod      
    
    cdate = str(yy)+"-"+str(doy)+"-"+str(datetime.timedelta(seconds=sod))
    fecha = datetime.datetime.strptime(cdate, "%y-%j-%X")
    
    return fecha  

# Convert and assign
m = df3.apply(change2datetime, axis=1)

# Create the new column
df3.insert(0,column="DateTime",value=0)
df3["DateTime"] = m

# Delete some columns 
del df3["YY"]
del df3["DOY"]
del df3["SOD"]

df3
```
