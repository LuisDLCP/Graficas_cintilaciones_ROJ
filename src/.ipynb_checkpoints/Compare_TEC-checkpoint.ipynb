```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df1 = pd.read_csv("septentrio.csv", parse_dates=["DateTime"])
df2 = pd.read_csv("colorado.csv", parse_dates=["DateTime"])
```

```python
#df1.set_index("DateTime",inplace=True)
#df2.set_index("DateTime",inplace=True)
```

```python
for prn in df1["PRN"].unique():
    mask1 = df1["PRN"]==prn
    aux1 = df1[mask1]
    
    mask2 = df2["PRN"]==prn
    aux2 = df2[mask2]
    
    x1=aux1["DateTime"]
    y1=aux1["sTEC"]

    x2=aux2["DateTime"]
    y2=aux2["sTEC"]
    
    fig,ax = plt.subplots()
    locator = mdates.AutoDateLocator()
    formatter = mdates.ConciseDateFormatter(locator)
    ax.xaxis.set_major_formatter(formatter)#(mdates.DateFormatter("%H:%M"))
    ax.scatter(x1,y1,marker="o", label="Sept",color="limegreen")#"darkorange")
    ax.scatter(x2,y2,marker="o",label="Col",color="blue")
    ax.set(xlabel="Time", ylabel="sTEC", title=f"sTEC - PRN {prn}")
    ax.grid()
    plt.legend()
    plt.savefig(f"sTEC_PRN{prn}.png")
       
```
